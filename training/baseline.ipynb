```python
# Importing required libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

%matplotlib inline
```

```python
# path to root directory
path = Path('day_night_data/day_night_dataset/')
```

```python
# paths of train and valid day/night images
train_day = list((path/'train/day').iterdir())
train_night = list((path/'train/night').iterdir())
valid_day = list((path/'val/day').iterdir())
valid_night = list((path/'val/night').iterdir())
```

## EDA

Let's see some sample day and night images

```python
img_day = cv2.imread(str(train_day[0]))
img_night = cv2.imread(str(train_night[0]))
```

```python
# Resizing image to height and width of 500
img_day = cv2.resize(img_day, (500,500))
img_night = cv2.resize(img_night, (500,500))
```

```python
# Converting from BGR format to RGB format for visualization
day_rgb = cv2.cvtColor(img_day, cv2.COLOR_BGR2RGB)
night_rgb = cv2.cvtColor(img_night, cv2.COLOR_BGR2RGB)
```

```python
# Visualizing images using matplotlib
fig, ax = plt.subplots(1,2,figsize=(10,15))
ax[0].imshow(day_rgb)
ax[0].set_title('Day')
ax[1].imshow(night_rgb)
ax[1].set_title('Night')
```

Images taken during day are generally brighter than images taken during night. We can use this fact to build a simple baseline model.
For this, we need to get the average brightness in an image.
RGB image does not help much in this case.
We can convert the image from RGB colorspace to Hue Saturation Value (HSV) colorspace.
The Value in HSV indicates brightness at different positions. We can utilize image from this colorspace to build a basic classifier.

```python
# converting image to HSV colorspace
day_hsv = cv2.cvtColor(img_day, cv2.COLOR_BGR2HSV)
night_hsv = cv2.cvtColor(img_night, cv2.COLOR_BGR2HSV)
```

```python
# Visualizing images using matplotlib
fig, ax = plt.subplots(1,2,figsize=(10,15))
ax[0].imshow(day_hsv)
ax[0].set_title('Day')
ax[1].imshow(night_hsv)
ax[1].set_title('Night')
```

This does not look useful. Let's split the channels and visualize them.

```python
# splitting channels of day and night hsv images
dh, ds, dv = cv2.split(day_hsv)
nh, ns, nv = cv2.split(night_hsv)
```

```python
fig, ax = plt.subplots(2,3,figsize=(15,10))
ax[0][0].imshow(dh)
ax[0][0].set_title('Hue')
ax[0][1].imshow(ds)
ax[0][1].set_title('Saturation')
ax[0][2].imshow(dv)
ax[0][2].set_title('Value')

ax[1][0].imshow(nh)
ax[1][0].set_title('Hue')
ax[1][1].imshow(ns)
ax[1][1].set_title('Saturation')
ax[1][2].imshow(nv)
ax[1][2].set_title('Value')
```

The value channel has higher pixel values where image is bright.
For example, the sun rays in day image and fireworks in night image have higher pixel values.

## Baseline model

Let's try to find average brightness of day and night images. We can use this as threshold to classify images.

```python
# arrays to store avg. brightness from Value channel of each image
day_brightness = []
night_brightness = []
```

```python
for curr_file in train_day:
  img = cv2.imread(str(curr_file)) # reading img 
  img = cv2.resize(img, (500, 500)) # resizing image
  img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) # converting to hsv
  avg_brightness = np.mean(img[:, :, 2]) # calculating average value of Value channel from HSV image
  day_brightness.append(avg_brightness) # appending to array
```

```python
for curr_file in train_night:
  img = cv2.imread(str(curr_file)) # reading img 
  img = cv2.resize(img, (500, 500)) # resizing image
  img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) # converting to hsv
  avg_brightness = np.mean(img[:, :, 2]) # calculating average value of Value channel from HSV image
  night_brightness.append(avg_brightness) # appending to array
```

```python
# calculating average brightness
day_avg_brightness = sum(day_brightness)/len(day_brightness)
night_avg_brightness = sum(night_brightness)/len(night_brightness)
```

```python
day_avg_brightness, night_avg_brightness
```

Let's try to visualize distribution of brightness in day and night images

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].hist(day_brightness)
ax[0].set_title('Day')
ax[1].hist(night_brightness)
ax[1].set_title('Night')
```

If we use a threshold=100 for average brightness, we can see that it covers much of the distribution for Day images as well as Night images

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].hist(day_brightness)
ax[0].set_title('Day')
ax[0].axvline(100, color='red')
ax[1].hist(night_brightness)
ax[1].set_title('Night')
ax[1].axvline(100, color='red')
```

## Validation

Let's write a simple function that takes in threshold as input, classifies images in validation set and returns the accuracy

```python
def validate(thresh=100):

  corrects = 0 # tracks running correct values
  total = len(valid_day) + len(valid_night) # total number of images in validaton set

  for curr_file in valid_day:
    img = cv2.imread(str(curr_file)) # reading image
    img = cv2.resize(img, (500,500)) # resizing image to standard size
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) # converting to HSV
    avg_brightness = np.mean(img[:, :, 2]) # calculating average value of Value channel in HSV image
    if avg_brightness > thresh:
      corrects += 1 # To classify image as Day, the avg. brightness should be greater than threshold brightness
    
  for curr_file in valid_night:
    img = cv2.imread(str(curr_file)) # reading image
    img = cv2.resize(img, (500,500)) # resizing image to standard size
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) # converting to HSV
    avg_brightness = np.mean(img[:, :, 2]) # calculating average value of Value channel in HSV image
    if avg_brightness < thresh:
      corrects += 1 # To classify image as Night, the avg. brightness should be less than threshold brightness
  
  accuracy = (corrects * 1.0)/total # calculating percentage of correctly classified images
  return accuracy
```

Let's try to validate for threshold = 100

```python
validate(thresh=100)
```

We got an Accuracy of 88.5% with this approach!

Now, let's try to decrease and increase threshold

```python
validate(thresh=90)
```

```python
validate(thresh=110)
```

Threshold of 100 seems like the sweet spot to achieve maximum accuracy.
